```python
import random
import numpy as np
import shutil
import os
random.seed(1337)
np.random.seed(1337)

#scan folders
video = 2 # 1 or 2
test_ratio = 0.2
path_imgs = '/content/drive/MyDrive/Final Project/project_20_data/video' + str(video) + '/images/'
path_xmls = '/content/drive/MyDrive/Final Project/project_20_data/video' + str(video) + '/frames/'
xmls = list(sorted(os.listdir(path_xmls)))
np.random.shuffle(xmls)

xmls_no = len(xmls)
xmls_test_no = int(xmls_no*test_ratio)
xmls_test = xmls[:xmls_test_no]
xmls_train = xmls[xmls_test_no:]

test_dir = '/content/drive/MyDrive/Final Project/project_20_data/video' + str(video) + '/test/'
train_dir = '/content/drive/MyDrive/Final Project/project_20_data/video' + str(video) + '/train/'
xmls_test_dir = test_dir + 'frames/'
xmls_train_dir = train_dir + 'frames/'
imgs_test_dir = test_dir + 'images/'
imgs_train_dir = train_dir + 'images/'

if not os.path.exists(test_dir):
  os.mkdir(test_dir)
  os.mkdir(xmls_test_dir)
  os.mkdir(imgs_test_dir)
  for f in xmls_test:
    shutil.copy(path_xmls + f, xmls_test_dir)
    name_base, _ = os.path.splitext(f)
    shutil.copy(path_imgs + name_base + '.PNG', imgs_test_dir)

  os.mkdir(train_dir)
  os.mkdir(xmls_train_dir)
  os.mkdir(imgs_train_dir)
  for f in xmls_train:
    shutil.copy(path_xmls + f, xmls_train_dir)
    name_base, _ = os.path.splitext(f)
    shutil.copy(path_imgs + name_base + '.PNG', imgs_train_dir)
```
