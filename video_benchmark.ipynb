```python
import sys
sys.path.append('/content/drive/MyDrive/Final Project/')
from functions import get_transform, save_model, load_model, convert_model_backbone
from ColaBeerDataset import ColaBeerDataset
from benchmarking import benchmark_record
from ModelResnet50 import ModelResnet50
from ModelMobilenetV2 import ModelMobilenetV2, ModelMobilenetV2_int8
from ModelMobilenetV3 import ModelMobileNetV3, ModelMobileNetV3L, ModelMobileNetV3L_int8
from ModelSSD import ModelSSDlite320

import numpy as np
import torch
from PIL import Image, ImageDraw
import cv2
```

```python
path_video = '/content/drive/MyDrive/Final Project/project_20_data/video'
video_no = 2 #1 or 2
path_video += str(video_no)
dataset = ColaBeerDataset(path_video,get_transform(train=False))
```

```python
model_resnet50 = ModelResnet50.get_model(3)
load_model("bbox_detector_resnet_e30_lr-4_g2_adam", model_resnet50)
benchmark_record('FRCNN_resnet50', model_resnet50, dataset)
```

```python
model_mobilenetv2 = ModelMobilenetV2.get_model(3)
load_model("bbox_detector_mobilenetv2_e30_lr-4_g2_adam", model_mobilenetv2)
benchmark_record('FRCNN_mobilenetv2', model_mobilenetv2, dataset)
```

```python
model_mobilenetv2 = ModelMobilenetV2.get_model(3)
model_mobilenetv2.half()
for layer in model_mobilenetv2.modules():
  if isinstance(layer, torch.nn.BatchNorm2d):
    layer.float()

load_model("bbox_detector_mobilenetv2_half_e30_lr-4_g2_adam", model_mobilenetv2)
benchmark_record('FRCNN_mobilenetv2_f16', model_mobilenetv2, dataset, use_half = True)
```
